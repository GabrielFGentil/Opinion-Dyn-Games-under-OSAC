# Using Erdos-Renyi

For Figure 11 and 12, we use Erdos-Renyi random graph. Cell bellow generate the graph.

```julia
nodes = 1000
g = Graphs.erdos_renyi(nodes, 300000, is_directed=true)
println("Connected: ", Graphs.is_connected(g), ". Strongly Connected: ", Graphs.is_strongly_connected(g), ".")
```

Next cell uses _rs_ise_ to turn _g_ in row stochastic matrix.

```julia
A = Graphs.adjacency_matrix(g);
A = rs_ize(A);
```

Next cell sets initial conditions, players targets and setting needed for the control function.

```julia
N_a = size(A,1)
x_init = rand(Uniform(-1,1),N_a,1)
γ = [0.001; 0.001];

N_p = 2
x_1_hat = 1 #Player 1 is the president
x_2_hat = -1 #Player 2 is the instructor

targets = zeros(N_a,N_p)

n_sample = 450

targets_1 = sample(1:Int(nodes/2), n_sample, replace = false)
targets_2 = sample(Int(nodes/2)+1:nodes, n_sample, replace = false)

targets_conflict = []
targets_1_ex = []
targets_2_ex = []
targets_not_target = []

for i in targets_1
   if i in targets_2
        push!(targets_conflict,i)
    else
        push!(targets_1_ex,i)
    end
end
k = 0
for i in targets_2
   if i in targets_1
        k = k+1
    else
        push!(targets_2_ex,i)
    end
end

for i in 1:nodes
   if i ∉ targets_1 &&  i ∉ targets_2
        push!(targets_not_target,i)
    end
end

for i in targets_1#Agents that the president is trying to convince to take his side
    targets[i,1] = x_1_hat
end

#Targets for the instructor
for i in  targets_2#Agents that the instructor is trying to convince to take his side
    targets[i,2] = x_2_hat
end

B = [targets[:,1]./x_1_hat targets[:,2]./x_2_hat];

T = 60; #horizon

#Setting for FJ and HK
Θ = 0; #Just to call function
F = 0; #Just to call function
W = 0; #Just to call function

proj_flag = 0; #Projection flag = 1, all opinios are projected in a [-1;1] opinions
perm_flag = 1;
model = 1 #Recalling that: 1 - DeGroot; 2 - FJ; 3 - HK

u_min = -1;
u_max = 1;
x_os, u_os = COD_OSA_GS(x_init,targets,γ,T,A,B,Θ,F,W,proj_flag,perm_flag, model);
```

# Fig 11

```julia
figure(figsize=(6,6))

PyPlot.plot(x_os[targets_1_ex,1:end]',"*-")
PyPlot.plot(x_os[targets_2_ex,1:end]',"*-")

PyPlot.grid()
title("Time plots of opinion")
xlabel("Time")
ylabel(L"Opinions $x_i$")

#PyPlot.savefig("ER_players_plot.eps")
```

# Fig 12

```julia
figure(figsize=(6,6))
PyPlot.plot(x_os[targets_not_target,1:end]',"*-")

PyPlot.grid()
title("Time plots of opinion")
xlabel("Time")
ylabel(L"Opinions $x_i$")

#PyPlot.savefig("ER_notarget_plot.eps")
```

# Control plot

```julia
PyPlot.plot(u_os',"*-")

PyPlot.grid()
title("Control")
xlabel("Time")
ylabel(L"Controls $u_i$")
legend([L"$u_1$", L"$u_2$"])
```
